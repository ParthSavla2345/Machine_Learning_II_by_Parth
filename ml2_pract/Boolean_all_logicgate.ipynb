```python
import numpy as np

def step(x):
    return 1 if x>0 else 0


def perceptron_training(X, y, lr= 0.1, epochs= 20):
    w = np.zeros(X.shape[1])
    b = 0
    for epoch in range(epochs):
        for i in range(len(X)):
            z = np.dot(X[i], w) + b
            y_pred = step(z)
            error = y[i] - y_pred
            w += lr * error * X[i]
            b += lr * error
    return w,b

def prediction(X,w,b):
    return [step(np.dot(x,w) + b) for x in X]
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y_and = np.array([0,0,0,1])

w_and, b_and = perceptron_training(X,y_and)
pred = prediction(X,w_and,b_and)
print("Weights are : ",w_and)
print("Bias are : ",b_and)
print("X  Y | Output")
for i in range(len(X)):
    print(f"{X[i][0]}  {X[i][1]} | {pred[i]}")

```

```python
y_or = np.array([0,1,1,1])

w_or,b_or = perceptron_training(X,y_or)
pred = prediction(X,w_or,b_or)

print(f"Final weights are : {w_or}")
print(f"Final bias are : {b_or}")
print("X  Y | Output")
for i in range(len(X)):
    print(f"{X[i][0]}  {X[i][1]} | {pred[i]}")
```
